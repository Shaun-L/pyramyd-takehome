```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pandas as pd

# Load SBERT model once
model = SentenceTransformer("all-MiniLM-L6-v2")

def get_similar_vendors(df, software_category, query, threshold=0.4):
    processed = []

    for idx, row in df.iterrows():
        # Filter only vendors in the target category
        if software_category not in row["categories"]:
            continue

        # Handle missing values
        description = row["description"] if pd.notna(row["description"]) else ""
        overview = row["overview"] if pd.notna(row["overview"]) else ""
        categories = " ".join(eval(row["categories"])) if isinstance(row["categories"], str) else ""

        # Combine fields with weights
        full_text = (categories + " ") * 3 + (overview + " ") * 2 + description

        processed.append({
            "product_name": row["product_name"],
            "text": full_text.strip()
        })

    if not processed:
        return []

    # Extract text and encode
    vendor_texts = [entry["text"] for entry in processed]
    vendor_embeddings = model.encode(vendor_texts, convert_to_tensor=True)
    query_embedding = model.encode(query, convert_to_tensor=True)

    # Compute cosine similarities
    similarities = cosine_similarity([query_embedding], vendor_embeddings)[0]

    # Attach scores and filter
    for i, sim in enumerate(similarities):
        processed[i]["similarity"] = sim

    filtered = [entry for entry in processed if entry["similarity"] >= threshold]

    # Sort and return top N
    top_vendors = sorted(filtered, key=lambda x: x["similarity"], reverse=True)
    return top_vendors
```

```python
vendors = get_similar_vendors(df, "CRM", "email automation and lead tracking")
for v in vendors:
    print(v["product_name"], f"score: {v['similarity']:.2f}")
```
